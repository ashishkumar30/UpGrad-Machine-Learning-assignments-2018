```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_digits
```

```python
df=load_digits()
```

```python
x=df.data
y=df.target
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 10)
```

```python
pred = km.fit_predict(x)
```

```python
lbl = np.zeros_like(pred)
```

```python
from scipy.stats import mode
```

```python
for i in range(10):
    mask = (pred==i)
    lbl[mask]=mode(y[mask])[0]
```

```python
result = pd.DataFrame({'Labels':list(lbl),'Predict':list(pred)})
result.head(10)
```

```python
import sklearn.metrics as smt
```

```python
print('Accuracy score : ',smt.accuracy_score(y,lbl))
```

```python
print('Confusion matrix :\n',smt.confusion_matrix(y,lbl))
```

```python
x,y,z = img.shape
img2 = img.reshape(x*y,z)
img2.shape
```

```python
km = KMeans(n_clusters = 7)
km.fit_predict(img2)
cluster_centers = km.cluster_centers_
cluster_centers
```

```python
cluster_labels = km.labels_
cluster_labels
```

```python
pic = cluster_centers[cluster_labels]
pic
```
